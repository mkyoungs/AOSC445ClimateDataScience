## 1. Import Libraries

These imports will be used throughout the notebook.`cartopy.crs` contains coordinate reference systems (projections).`cartopy.feature` contains map layers like borders and coastlines.

```python
import pandas as pd
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
```

## 2. Read in data

```python
url = "https://www1.ncdc.noaa.gov/pub/data/ghcn/daily/ghcnd-stations.txt"
colspecs = [    (0,11),    (12,20),    (21,30),    (31,37),    (38,40),    (41,71)]
names = [    "station_id",    "latitude",    "longitude",    "elevation",    "state",    "name"]
stations = pd.read_fwf(url, colspecs=colspecs, names=names)
stations.head()
```

## 3. Filter stations

```python
us_stations = stations[stations["station_id"].str.startswith("US")]
len(us_stations)
```

## 4. Map Stations — PlateCarree Projection

Plot the station locations on the map with borders and state lines for the US stations.

**PlateCarree** is the simplest projection.It maps latitude directly to y and longitude to x.Most geospatial datasets are already in this coordinate system.

```python
fig = plt.figure(figsize=(12,7))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.set_extent([-125,-65,24,50])
ax.coastlines()
ax.add_feature(cfeature.BORDERS)
ax.add_feature(cfeature.STATES)
ax.scatter(us_stations["longitude"],us_stations["latitude"],s=8,transform=ccrs.PlateCarree())
plt.title("NOAA Weather Stations — PlateCarree")
plt.show()
```

## 6. Mercator Projection

Now plot using the Mercator projection for North America.

Mercator projection preserves **angles**, making it useful for navigation.However, it distorts area at high latitudes.

```python
fig = plt.figure(figsize=(12,7))
ax = plt.axes(projection=ccrs.Mercator())
ax.set_extent([-125,-65,24,50], crs=ccrs.PlateCarree())
ax.coastlines()
ax.add_feature(cfeature.STATES)
ax.scatter(stations["longitude"],stations["latitude"],s=8,transform=ccrs.PlateCarree())
plt.title("Weather Stations — Mercator Projection")
plt.show()
```

## 7. Lambert Conformal Projection

Now plot using the Lambert Conformal projection for the US

Lambert Conformal is widely used in **meteorology** for regional weather maps.It preserves shape locally across mid‑latitude regions like the United States.

```python
fig = plt.figure(figsize=(12,7))
ax = plt.axes(projection=ccrs.LambertConformal())
ax.set_extent([-125,-65,24,50], crs=ccrs.PlateCarree())
ax.coastlines()
ax.add_feature(cfeature.STATES)
ax.scatter(    stations["longitude"],    stations["latitude"],    s=8,    transform=ccrs.PlateCarree())
plt.title("Weather Stations — Lambert Conformal")
plt.show()
```

## 8. Robinson Projection

Try the same with the Robinson Projection

Robinson projection is commonly used for **global maps**. It balances distortion of area and shape.

```python
fig = plt.figure(figsize=(12,7))
ax = plt.axes(projection=ccrs.Robinson())
ax.set_global()
ax.coastlines()
ax.scatter(    stations["longitude"],    stations["latitude"],    s=8,    transform=ccrs.PlateCarree())
plt.title("Weather Stations — Robinson Projection")
plt.show()
```
