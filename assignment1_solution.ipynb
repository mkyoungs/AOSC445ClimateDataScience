### 1. Write a function to convert temperature from Kelvin to Celsius and Celsius to Kelvin
(2 pt)

````python
def CKTempConvert(temp,tounit):
    # function to convert kelvin to celsius or vice versa
    # tounit = 'C' means convert to celsius, 'K', to kelvin
    if tounit == 'C':
        tout = temp - 273.15
    elif tounit == 'K':
        tout = temp + 273.15
    else:
         raise ValueError("improper flag: use 'C' or 'K'")
    return tout
````

### 6. Write a function takes one argument (height in meters) and returns two arguments (feet and inches)

(Consult the [tutorial on numbers](https://docs.python.org/3/tutorial/introduction.html#numbers) if you are stuck on how to implement this.)
(3 pt)

````python
def toFtInch(meters):
    inches = meters*39.3701
    ft = inches//12
    inch = inches%12
    return ft, inch
````
